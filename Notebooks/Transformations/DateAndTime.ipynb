# Build Spark Session

```python
import os
import sys

os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"

from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("DataRead")
    .master("local[*]")
    .getOrCreate()
)

spark
```

# Truncate –Year, Quarter, Month ,Week, Day, Hour, Minute, Second

```python
from pyspark.sql.functions import *
from pyspark.sql.types import StructType

dual_df = spark.createDataFrame([()], StructType([]))

dual_df.select(
    current_timestamp().alias("Current_Timestamp"),
    date_trunc("year", current_timestamp()).alias("Start_of_Year"),
    date_trunc("quarter", current_timestamp()).alias("Start_of_Quarter"),
    date_trunc("month", current_timestamp()).alias("Start_of_Month"),
    date_trunc("week", current_timestamp()).alias("Start_of_Week"),
    date_trunc("day", current_timestamp()).alias("Start_of_Day"),
    date_trunc("hour", current_timestamp()).alias("Start_of_Hour"),
    date_trunc("minute", current_timestamp()).alias("Start_of_Minute"),
    date_trunc("second", current_timestamp()).alias("Start_of_Second")
).show(truncate=False)
```

### Extracts –Year, Quarter, Month ,Week, Day, Hour, Minute, Second

```python
from pyspark.sql.functions import *
from pyspark.sql.types import StructType

dual_df = spark.createDataFrame([()], StructType([]))

dual_df.select(
    current_timestamp().alias("Current_Timestamp"),
    year(current_timestamp()).alias("Year"),
    quarter(current_timestamp()).alias("Quarter"),
    month(current_timestamp()).alias("Month"),
    weekofyear(current_timestamp()).alias("Week_of_Year"),
    dayofyear(current_timestamp()).alias("Day_of_Year"),
    dayofmonth(current_timestamp()).alias("Day_of_Month"),
    dayofweek(current_timestamp()).alias("Day_of_Week"),
    hour(current_timestamp()).alias("Hour"),
    minute(current_timestamp()).alias("Minute"),
    second(current_timestamp()).alias("Second"),
    last_day(current_timestamp()).alias("Last_Day_of_Month")
).show(truncate=False)
```

### Addition –Year, Quarter, Month ,Week, Day, Hour, Minute, Second

```python
from pyspark.sql.functions import *
from pyspark.sql.types import StructType

dual_df = spark.createDataFrame([()], StructType([]))

dual_df.select(
    current_timestamp().alias("Current_Timestamp"),
    date_add(current_timestamp(), 2).alias("Date_After_2_Days"),
    date_add(current_timestamp(), -2).alias("Date_Before_2_Days"),
    add_months(current_timestamp(), 2).alias("Date_After_2_Months"),
    add_months(current_timestamp(), -2).alias("Date_Before_2_Months")
).show(truncate=False)
```

### Addition –Year, Quarter, Month ,Week, Day, Hour, Minute, Second

```python
from pyspark.sql.functions import *
from pyspark.sql.types import StructType

dual_df = spark.createDataFrame([()], StructType([]))

dual_df.select(
    current_timestamp().alias("Current_Timestamp"),
    (current_timestamp() + expr("INTERVAL 2 DAYS")).alias("Date_After_2_Days"),
    (current_timestamp() - expr("INTERVAL 2 DAYS")).alias("Date_Before_2_Days"),
    (current_timestamp() + expr("INTERVAL 2 MONTHS")).alias("Date_After_2_Months"),
    (current_timestamp() - expr("INTERVAL 2 MONTHS")).alias("Date_Before_2_Months"),
    (current_timestamp() + expr("INTERVAL 2 YEARS")).alias("Date_After_2_Years"),
    (current_timestamp() - expr("INTERVAL 2 YEARS")).alias("Date_Before_2_Years"),
    (current_timestamp() + expr("INTERVAL 2 HOURS")).alias("Date_After_2_Hours"),
    (current_timestamp() - expr("INTERVAL 2 HOURS")).alias("Date_Before_2_Hours"),
    (current_timestamp() + expr("INTERVAL 2 MINUTES")).alias("Date_After_2_Minutes"),
    (current_timestamp() - expr("INTERVAL 2 MINUTES")).alias("Date_Before_2_Minutes"),
    (current_timestamp() + expr("INTERVAL 2 SECONDS")).alias("Date_After_2_Seconds"),
    (current_timestamp() - expr("INTERVAL 2 SECONDS")).alias("Date_Before_2_Seconds")
).show(truncate=False)
```

### Difference –date, month

```python
from pyspark.sql.functions import *
from pyspark.sql.types import StructType

dual_df = spark.createDataFrame([()], StructType([]))

dual_df.select(
    current_date().alias("Current_Date"),
    to_date(lit("2023-01-01")).alias("Old_Date"),
    date_diff(current_date(), to_date(lit("2023-01-01")), "days").alias("Days_Difference"),
    date_diff(current_date(), to_date(lit("2023-01-01")), "months").alias("Months_Difference")
).show(truncate=False)
```

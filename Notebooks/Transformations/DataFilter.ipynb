# Build Spark Session

```python
import os
import sys

os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"

from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("DataRead")
    .master("local[*]")
    .getOrCreate()
)

spark
```

### Remove Rows those are having null values in all columns

```python
customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Filter_Data\Products_Purchase_Data_20210111.csv")
)
print("Source Data with Null Values")
customer_df.show()

print("Source Data with out Null Values (All Columns are null  for a row)")
customer_df.na.drop(how="all").show()
```

### Remove Rows those are having null values in any column

### Identify rows at least two columns having valid value

```python
customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Filter_Data\Products_Purchase_Data_20210111.csv")
)
print("Source Data with Null Values")
customer_df.show()

print("Source Data atleast having 2 non-null values for a row")
customer_df.na.drop(how="any", thresh=2).show()
```

### Identify Rows Either Customer ID, country Code dont have null values

```python
customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Filter_Data\Products_Purchase_Data_20210111.csv")
)
print("Source Data with Null Values")
customer_df.show()

print("Source Data Either Customer Id or Country Code is not null for a row")
customer_df.na.drop(how="any", subset=["Customerid","CountryCode"]).show()
```

### Fill the Missing Values

```python
customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Filter_Data\Products_Purchase_Data_20210111.csv")
)
print("Source Data with Null Values")
customer_df.show()

print("Filling Missing Values with Default Zip Code as 999999 for Null Values in ZipCode Column")
customer_df.na.fill({"ZipCode": "999999"}).show()
```

### Filter the data where country code is null

```python
customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Filter_Data\Products_Purchase_Data_20210111.csv")
)
print("Source Data with Null Values")
customer_df.show()

print("Filtering out rows where country code is null")
customer_df.filter(customer_df.CountryCode.isNull()).show()
```

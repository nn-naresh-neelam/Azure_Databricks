# Build Spark Session

```python
import os
import sys

os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"

from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("DataRead")
    .master("local[*]")
    .getOrCreate()
)

spark
```

## Joins

### PySpark Key Word: { inner }, SQL Equivalent : INNER JOIN
### PySpark Key Word: { cross }, SQL Equivalent : CROSS JOIN
### PySpark Key Word: { outer, full, fullouter, full_outer }, SQL Equivalent : OUTER JOIN
### PySpark Key Word: { left, leftouter, left_outer }, SQL Equivalent : LEFT OUTER JOIN
### PySpark Key Word: { right, rightouter, right_outer }, SQL Equivalent : RIGHT OUTER JOIN
### PySpark Key Word: { semi, leftsemi, left_semi }, SQL Equivalent : same as INNER JOIN , but it will give data from base table for matching records
### PySpark Key Word: { anti, leftanti, left_anti }, SQL Equivalent : opposite of INNER JOIN , it will display non matching records from left table only.

## Source Data Set

```python
Customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Joins\customers.csv")
)

print("Customer Data")
Customer_df.show()

Orders_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(r"C:\Users\navat\AzureDataBricks\Azure_Databricks\Raw_Source\Joins\orders.csv")
)

print("Orders Data")
Orders_df.show()
```

### PySpark Key Word: { inner }, SQL Equivalent : INNER JOIN

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "inner").show()
```

### PySpark Key Word: { cross }, SQL Equivalent : CROSS JOIN

```python
Customer_df.crossJoin(Orders_df).show()
```

### PySpark Key Word: { outer, full, fullouter, full_outer }, SQL Equivalent : FULL OUTER JOIN

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "outer").show()
```

### PySpark Key Word: { left, leftouter, left_outer }, SQL Equivalent : LEFT OUTER JOIN

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "left").show()
```

### PySpark Key Word: { right, rightouter, right_outer }, SQL Equivalent : RIGHT OUTER JOIN

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "right").show()
```

### PySpark Key Word: { semi, leftsemi, left_semi }, SQL Equivalent : same as INNER JOIN , but it will give data from base table for matching records

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "semi").show()
```

### PySpark Key Word: { anti, leftanti, left_anti }, SQL Equivalent : opposite of INNER JOIN , it will display non matching records from base table only.

```python
Customer_df.join(Orders_df, Customer_df.customer_id == Orders_df.customer_id, "anti").show()
```
